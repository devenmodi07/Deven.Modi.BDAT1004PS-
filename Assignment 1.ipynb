```python
#Question 1


5 is an integer.
5.0 is a floating - point number.
5 > 1 is a boolean value, True.
'5' is a string.
5 * 2 is an integer, 10.
'5' * 2 is a string, '55'.
'5' + '2' is a string, '52'.
5 / 2 is a floating - point number, 2.5.
5 % 2 is an integer, 1.
{ 5, 2, 1} is a set.
5 == 3 is a boolean value, False.
Pi is a mathematical constant that can be represented as a floating-point number, for example 3.14159265358979323846.


#Question 2


string word1 = "Supercalifragilisticexpialidocious";
string word2 = "Honorificabilitudinitatibus";
string word3 = "Bababadalgharaghtakamminarronnkonn";
string[] composers = { "Berlioz", "Borodin", "Brian", "Bartok", "Bellini", "Buxtehude", "Bernstein" };

// a. How many letters are there in 'Supercalifragilisticexpialidocious'?
int letterCount = word1.Length;

// b. Does 'Supercalifragilisticexpialidocious' contain 'ice' as a substring?
bool containsIce = word1.Contains("ice");

// c. Which of the following words is the longest:
// Supercalifragilisticexpialidocious, Honorificabilitudinitatibus, or
// Bababadalgharaghtakamminarronnkonn?
string longestWord = word1.Length > word2.Length ? word1 : word2;
longestWord = longestWord.Length > word3.Length ? longestWord : word3;

// d. Which composer comes first in the dictionary: 'Berlioz', 'Borodin', 'Brian',
// 'Bartok', 'Bellini', 'Buxtehude', 'Bernstein'. Which one comes last?
Array.Sort(composers);
string firstComposer = composers[0];
string lastComposer = composers[composers.Length - 1];

// Output
Console.WriteLine("The word '" + word1 + "' has " + letterCount + " letters.");
Console.WriteLine("Does the word '" + word1 + "' contain 'ice' as a substring? " + containsIce);
Console.WriteLine("The longest word is: " + longestWord);
Console.WriteLine("The first composer in the dictionary is: " + firstComposer);
Console.WriteLine("The last composer in the dictionary is: " + lastComposer);


#Question 3 


using System;

namespace TriangleArea
{
    class Program
    {
        static void Main(string[] args)
        {
            Console.WriteLine(triangleArea(2, 2, 2));
        }

        static double triangleArea(double a, double b, double c)
        {
            double s = (a + b + c) / 2;
            double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            return area;
        }
    }
}



#Question 4


using System;

namespace OddAndEvenArrays
{
    class Program
    {
        static void Main(string[] args)
        {
            Console.Write("Input the number of elements to be stored in the array: ");
            int n = int.Parse(Console.ReadLine());

            int[] arr = new int[n];

            Console.WriteLine("Input {0} elements in the array:", n);
            for (int i = 0; i < n; i++)
            {
                Console.Write("element - {0}: ", i);
                arr[i] = int.Parse(Console.ReadLine());
            }

            int[] even = new int[n];
            int[] odd = new int[n];
            int evenIndex = 0;
            int oddIndex = 0;

            foreach (int num in arr)
            {
                if (num % 2 == 0)
                {
                    even[evenIndex++] = num;
                }
                else
                {
                    odd[oddIndex++] = num;
                }
            }

            Console.WriteLine("The Even elements are:");
            for (int i = 0; i < evenIndex; i++)
            {
                Console.Write("{0} ", even[i]);
            }
            Console.WriteLine();

            Console.WriteLine("The Odd elements are:");
            for (int i = 0; i < oddIndex; i++)
            {
                Console.Write("{0} ", odd[i]);
            }
            Console.WriteLine();
        }
    }
}


#Question 5


public static bool Inside(double x, double y, double x1, double y1, double x2, double y2)
{
    return x >= x1 && x <= x2 && y >= y1 && y <= y2;
}

```

```python
#Question-6

def pig(word):
    vowels = "aeiou"
    if word[0].lower() in vowels:
        return word.lower() + "way"
    else:
        return word[1:].lower() + word[0].lower() + "ay"
    
pig("enter")
pig("happy")
```

```python
#Question-7


def bldcount(name):
    blood_types = {
        "A": 0,
        "B": 0,
        "AB": 0,
        "O": 0,
        "OO": 0
    }

    with open(name, "r") as file:
        for line in file:
            blood_type = line.strip()
            blood_types[blood_type] += 1

    for blood_type, count in blood_types.items():
        if count == 0:
            print("There are no patients of blood type {}.".format(blood_type))
        elif count == 1:
            print("There is one patient of blood type {}.".format(blood_type))
        else:
            print("There are {} patients of blood type {}.".format(count, blood_type))
            
bldcount("bloodtype.txt")

```

```python
#Question 8 


import os

def curconv(currency, amount):
    # read the currencies.txt file and store the exchange rates in a dictionary
    exchange_rates = {}
    with open(os.path.join(os.path.dirname(__file__), 'currencies.txt')) as f:
        for line in f:
            currency, rate = line.strip().split()
            exchange_rates[currency] = float(rate)

    # convert the input currency to USD
    return amount * exchange_rates.get(currency, 1)

curconv("EUR", 100)
curconv("JPY", 100)
```

```python
#Question 10


def frequencies(text):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    result = [0] * 26
    text = text.lower()
    for char in text:
        if char in letters:
            result[letters.index(char)] += 1
    return result

frequencies("The quick red fox got bored and went home")
frequencies("apple")
```
